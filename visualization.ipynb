```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

def display_board(board):
    cmap = ListedColormap(["white", "blue"])

    plt.imshow(board, cmap=cmap, origin='upper', interpolation='nearest')
    plt.title("9x9 Binary Grid")
    plt.show()
```

```python
from blockulib.playing import ModelBasedLoop
play = ModelBasedLoop()
```

```python
data = play(num_games = 1, top_k = 7, temperature = 0.3)
```

```python
for i in range(1, len(data[0])):
    display_board(data[0][i]-data[0][i-1])
```

```python
for board in data[0]:
    display_board(board)
```
